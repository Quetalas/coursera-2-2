```python
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
circles = datasets.make_circles()
```

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['red','yellow'])
plt.figure()
plt.scatter(circles[0][:,0], circles[0][:,1], c=circles[1], cmap=colors)
```

```python
def plot_2d_dataset(data, colors):
    plt.figure()
    plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap=colors)
```

```python
noisy_circles = datasets.make_circles(noise=0.05)
plot_2d_dataset(noisy_circles, colors)
```

```python
noisy_circles = datasets.make_circles(noise=0.5)
plot_2d_dataset(noisy_circles, colors)
```

```python
simple_classification_problem = datasets.make_classification(n_features=2, n_informative=1, n_redundant=1, n_clusters_per_class=1, random_state=1)
```

```python
plot_2d_dataset(simple_classification_problem, colors)
```

```python
clasification_problem = datasets.make_classification(n_features=2, n_informative=2, n_classes=4, n_redundant=0, n_clusters_per_class=1, random_state=1)
colors = ListedColormap(['red', 'blue', 'green', 'yellow'])
```

```python
plot_2d_dataset(clasification_problem, colors)
```
